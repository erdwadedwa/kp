```python
#Usuwam puste wiersze (kategorie dla których nie ma danych)
BezPustychKolumnIPustychWierszy = BezPustychKolumn.dropna (axis=0, thresh=3)
```

Wybieram tylko dane ze zbioru "Barro-Lee":

```python
BarroLee = BezPustychKolumnIPustychWierszy[BezPustychKolumnIPustychWierszy['series_name'].str.contains("Barro")]
```

```python
#BarroLee.info()
```

```python
BarroLee = BarroLee.dropna( axis =1, how ='all')
```

```python
BarroLee = BarroLee.dropna( axis =0, thresh = 3 )
```

```python
#BarroLee.info()
```

```python
BarroLee
```

Wyodrębniam dane na temat edukacji pierwszego stopnia:

```python

AverageYearsOf_PrimSchooling = BarroLee[BarroLee['series_name'].str.contains("Barro-Lee: Average years of primary schooling")]
AverageYearsOf_PrimSchooling
```

Odwracam dataframe i "spłaszczam" multiindeks do pojedynczego indeksu

```python
AverageYearsOf_PrimSchooling = AverageYearsOf_PrimSchooling.pivot(index='country_name', columns='series_name')
AverageYearsOf_PrimSchooling.columns = [' '.join(col).strip() for col in AverageYearsOf_PrimSchooling.columns.values]
```

```python
AverageYearsOf_PrimSchooling
```

Średnia ilość lat otrzymanej edukacji pierwszego stopnia, w grupie osób w wieku od 20 do 24 (w momencie badania, kobiety i meżczyźni łącznie) - porównanie danych na rok 1985 i 2010.

```python
sns.distplot(AverageYearsOf_PrimSchooling['1985 Barro-Lee: Average years of primary schooling, age 20-24, total'], bins=20, label='1985',
            color='blue',fit_kws = {'color': 'blue'})
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 20-24, total'], bins=20, label='2010',
             color = 'green', fit_kws = {'color': 'green'})
plt.legend()
```

Średnia ilość lat otrzymanej edukacji pierwszego stopnia, w grupie kobiet w wieku powyżej 75 lat (w momencie badania) - porównanie danych na rok 1985 i 2010.

```python
sns.distplot(AverageYearsOf_PrimSchooling['1985 Barro-Lee: Average years of primary schooling, age 75+, female'], label='1985',
            color='yellow',fit_kws = {'color': 'yellow'})
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 75+, female'], label='2010',
             color = 'red', fit_kws = {'color': 'red'})
plt.legend()
```

```python
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 75+, female'], kde=False, hist= False, fit = stats.gamma, label='women',
            color='red',fit_kws = {'color': 'red'})
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 75+, total'], kde=False, hist= False, fit = stats.gamma, label='total',
             color = 'green', fit_kws = {'color': 'green'})
plt.legend()
```

Średnia ilość lat otrzymanej edukacji pierwszego stopnia prez osoby w wieku powyżej 75 lat (w momencie badania) - porównanie danych na rok 1985 i 2010, dla kobiet oraz dla kobiet i mężczyzn łacznie.

```python
sns.distplot(AverageYearsOf_PrimSchooling['1985 Barro-Lee: Average years of primary schooling, age 75+, total'], bins=40, kde=False, vertical=True, label='1985, total',
            color='blue',fit_kws = {'color': 'blue'})
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 75+, total'], bins=40, kde=False, vertical=True, label='2010, total',
             color = 'green', fit_kws = {'color': 'green'})
sns.distplot(AverageYearsOf_PrimSchooling['1985 Barro-Lee: Average years of primary schooling, age 75+, female'], bins=40, kde=False, vertical=True, label='1985, women',
            color='yellow',fit_kws = {'color': 'yellow'})
sns.distplot(AverageYearsOf_PrimSchooling['2010 Barro-Lee: Average years of primary schooling, age 75+, female'], bins=40, kde=False, vertical=True, label='2010, women',
             color = 'red', fit_kws = {'color': 'red'})
plt.legend()
```
