```python
fig = plt.figure()
plt.plot(zarobki[szczescie > 80], szczescie[szczescie > 80], 'o', color = 'red', label = 'Bardzo szczęśliwi', marker='.')
plt.plot(zarobki[(szczescie > 60) & (szczescie <= 80)], szczescie[(szczescie >60) & (szczescie <= 80)], 'p', color='blue', label = 'Raczej szczęśliwi')
plt.plot(zarobki[(szczescie > 40) & (szczescie <= 60)], szczescie[(szczescie > 40) & (szczescie <= 60)], 'd', color='orange', label='Raczej nieszczęśliwi')
plt.plot(zarobki[szczescie <= 40], szczescie[szczescie <= 40], '<', color='green', label='Bardzo nieszczęśliwi')
fig.suptitle('Szczęście a zarobki w złotych', fontsize=16)
plt.xlabel('Zarobki w złotych', fontsize=12)
plt.ylabel('Szczęście mierzone kwestionariuszem')
plt.ylim((0,105))
plt.xlim((0,14000))
plt.legend()
fig.savefig('s01_pd03.png')
plt.show()
```
